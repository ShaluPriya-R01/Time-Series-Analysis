```python
from datetime import datetime
```

```python
import yfinance as yf
```

```python
data = yf.download("PRICOLLTD.NS",start="2024-01-01")
```

```python
data
```

```python
#Simple Moving Average
```

```python
data['Open'].plot(figsize=(15,6))
```

```python
data["Open:10 days rolling"]=data["Open"].rolling(window=10,min_periods=1).mean()
data["Open:30 days rolling"]=data["Open"].rolling(window=30,min_periods=1).mean()
data["Open:50 days rolling"]=data["Open"].rolling(window=50,min_periods=1).mean()
```

```python
data
```

```python
data[["Open","Open:10 days rolling"]].plot(xlim=['2024-01-01','2024-12-31'],figsize=(15,7))
```

```python
data[["Open","Open:10 days rolling","Open:30 days rolling","Open:50 days rolling"]].plot(xlim=['2024-01-01','2024-12-31'],figsize=(15,7))
```

```python
#Expanding

data['Open'].expanding().mean().plot(figsize=(10,3))
```

```python
data['Open'].expanding().mean()
```

```python
data['Open'].expanding(10).mean().plot(figsize=(10,3))
```

```python
data['Open'].expanding(10).mean().plot(figsize=(10,3))
```

```python
#Exponential Moving Average
```

```python
data['EMA_0.1']=data['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
data
```

```python
data[["Open","EMA_0.1"]].plot(xlim=['2024-01-01','2024-12-31'],figsize=(15,7))
```

```python
data['EMA_0.3']=data['Open'].ewm(alpha=0.3,adjust=False).mean()
```

```python
data[["Open","EMA_0.3"]].plot(xlim=['2024-01-01','2024-12-31'],figsize=(15,7))
```

```python
data['EMA_0.5']=data['Open'].ewm(alpha=0.5,adjust=False).mean()
```

```python
data
```

```python
data['EMA_5Days']=data['Open'].ewm(span=5).mean()
```

```python
data[['Open','EMA_0.3','EMA_5Days']].plot(xlim=['2024-01-01','2024-12-31'],figsize=(15,6))
```
