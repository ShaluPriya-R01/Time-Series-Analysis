# Time Resampling

```python
df_tesla.head()
```

```python
df_tesla.resample(rule="A")
```

```python
df_tesla.resample(rule='A').max()
```

```python
df_tesla.resample(rule="M")
```

```python
df_tesla.resample(rule="ME").max()
```

```python
df_tesla.resample(rule="W").min()
```

```python
df_tesla.resample(rule="YE")
```

```python
df_tesla.resample(rule="YE").min()
```

```python
df_tesla.resample(rule="M").max()['Open'].plot()
```

```python
df_tesla.resample(rule="QS").max()["High"].plot()
```

```python
df_tesla.resample(rule="BQ").max()["High"].plot()
```

```python
df_tesla["Open"].resample(rule="BQ").mean().plot(kind='bar')
```

```python
df_tesla.resample(rule="M").mean()["High"].plot()
```
